```python
import matplotlib.pyplot as plt
import cv2

images = [
    "/content/runs/detect/predict/download.jpg",   # YOLOv8
    "/content/runs/detect/predict3/download.jpg",  # YOLOv9
    "/content/runs/detect/predict5/download.jpg",  # YOLOv10
    "/content/runs/detect/predict6/download.jpg"   # YOLOv11
]
names = ["YOLOv8", "YOLOv9", "YOLOv10", "YOLOv11"]

plt.figure(figsize=(20, 5))
for i in range(4):
    img = cv2.imread(images[i])
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.subplot(1, 4, i+1)  # 1 row, 4 columns
    plt.imshow(img)
    plt.title(names[i])
    plt.axis("off")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
models = ["YOLOv8", "YOLOv9", "YOLOv10", "YOLOv11"]
inference_times = [195.5, 1791.1, 171.9, 109.7]  # in ms
mAPs = [0.995, 0.995, 0.992, 0.994]              # mAP@0.5
sizes = [5.96, 111.80, 5.48, 5.22]               # Model size in MB
colors = ["blue", "green", "pink", "red"]

# Create scatter plot with size
plt.figure(figsize=(8, 6))
for i in range(len(models)):
    plt.scatter(inference_times[i], mAPs[i], s=sizes[i]*50, color=colors[i], label=models[i], alpha=0.6)

# Label each point
for i, model in enumerate(models):
    plt.text(inference_times[i], mAPs[i], model, fontsize=10, ha='right', color=colors[i])

# Axis labels and title
plt.xlabel("Inference Time (ms)")
plt.ylabel("mAP@0.5")
plt.title("Performance Comparison of YOLO Models (Size Proportional to Model Size)")

# Adjust axis limits
plt.xlim(0, 2000)
plt.ylim(0.99, 1.0)

# Add legend
plt.legend()

# Save
plt.grid(True, linestyle='--', alpha=0.7)
plt.savefig("colored_size_comparison_graph.png", dpi=300, bbox_inches='tight')
plt.show()
```
